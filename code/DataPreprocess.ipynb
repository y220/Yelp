# Data Merge

```python
df_raw = pd.merge(review, business, how='left', on='business_id')
df_raw.reset_index(drop=True, inplace=True)
df_raw.head()
```

```python
df_raw.shape
```

```python
df_raw.columns
```

```python
df = df_raw.drop(columns=['user_id', 'useful', 'funny',
       'cool','date', 'address', 'postal_code', 'latitude', 'longitude', 'review_count',
       'is_open'])
df.shape
```

```python
categories = list(df.categories)
categories=[str(a) for a in categories]
categories=[a.lower() for a in categories]
categories_drop = [i for i in range(0,len(categories)) if categories[i].find('pizza')==-1 ]
```

```python
## Number of selected pizza review 
len(categories)-len(categories_drop)
```

```python
pizza=df.drop(labels=categories_drop,axis=0)
```

```python
## Number of selected pizza bussiness
len(set(list(pizza.business_id)))
```

```python
pizza=pizza.drop(columns=['business_id'])
pizza.reset_index(drop=True, inplace=True)
pizza.tail()
```

```python
pizza.to_csv('pizza_raw.csv')
```
